```python
import json


json_file = '/content/updated_cocktails.json'
output_json_file = 'output_cocktail_attributes.json'
attributes = [
    {"id": 1, "name": "savory", "category": "flavor"},
    {"id": 2, "name": "sweet", "category": "flavor"},
    {"id": 3, "name": "bitter", "category": "flavor"},
    {"id": 4, "name": "fruity", "category": "flavor"},
    {"id": 5, "name": "spicy", "category": "flavor"},
    {"id": 6, "name": "sour", "category": "flavor"},
    {"id": 7, "name": "smoky", "category": "flavor"},
    {"id": 8, "name": "herbaceous", "category": "flavor"},
    {"id": 9, "name": "salty", "category": "flavor"},
    {"id": 10, "name": "Romantic", "category": "occasion"},
    {"id": 11, "name": "Business", "category": "occasion"},
    {"id": 12, "name": "Celebration", "category": "occasion"},
    {"id": 13, "name": "Social", "category": "occasion"},
    {"id": 14, "name": "Summer", "category": "season"},
    {"id": 15, "name": "Winter", "category": "season"},
]


def assign_attributes(cocktail):
    assigned_attributes = []


    ingredients = ' '.join(cocktail['ingredients']).lower()


    if any(keyword in ingredients for keyword in ["sugar", "syrup", "cream", "honey"]):
        assigned_attributes.append({"cocktail_id": cocktail["id"], "attribute_id": 2, "value": True})


    if any(keyword in ingredients for keyword in ["fruit", "orange", "pineapple", "apple", "lime", "lemon", "cherry"]):
        assigned_attributes.append({"cocktail_id": cocktail["id"], "attribute_id": 4, "value": True})


    if any(keyword in ingredients for keyword in ["bitters", "campari"]):
        assigned_attributes.append({"cocktail_id": cocktail["id"], "attribute_id": 3, "value": True})

    if any(keyword in ingredients for keyword in ["lime", "lemon", "vinegar"]):
        assigned_attributes.append({"cocktail_id": cocktail["id"], "attribute_id": 6, "value": True})

    if cocktail['name'].lower() in ['margarita', 'mojito', 'pina colada']:
        assigned_attributes.append({"cocktail_id": cocktail["id"], "attribute_id": 13, "value": True})

    if any(keyword in cocktail['name'].lower() for keyword in ["mojito", "pina colada", "margarita"]):
        assigned_attributes.append({"cocktail_id": cocktail["id"], "attribute_id": 14, "value": True})

    if any(keyword in cocktail['name'].lower() for keyword in ["white russian", "irish coffee"]):
        assigned_attributes.append({"cocktail_id": cocktail["id"], "attribute_id": 15, "value": True})

    return assigned_attributes

def generate_cocktail_attributes(cocktails):
    cocktail_attributes = []

    for cocktail in cocktails:
        cocktail_attributes += assign_attributes(cocktail)

    return cocktail_attributes

with open(json_file, 'r') as f:
    cocktails_data = json.load(f)

cocktail_attributes = generate_cocktail_attributes(cocktails_data)

with open(output_json_file, 'w') as f:
    json.dump(cocktail_attributes, f, indent=4)

print(f"Cocktail attributes JSON saved to {output_json_file}")
```
